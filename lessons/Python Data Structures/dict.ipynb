Dictionary is a collection of unordered, modifiable(mutable) paired (key: value) data type

Making Dictionaries
    empty_dct = {}

    dct = {'key1':'value1', 'key2':'value2', 'key3':'value3'}

    dct = dict(key1='value1', key2='value2')

len(dct) - checks the number of the 'key:value' pairs

Accessing items
    dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
    print(dct['key1']) # value1
    print(dct['key4']) # value4

    get method returns None, which is a NoneType object data type, if the key does not exist unlike using the key cause which raises an error if the key does not exist 
    
    print(dct.get('key1'))

Adding items 
    dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
    dct['key5'] = 'value5'
    OR use update()
    upd_dict_values = {'key 5':'value 5'}
    dct.update(upd_dict_values)

Modifying values 
    dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
    dct['key1'] = 'value-one'

Checking keys, we use the in operator
    print('key2' in dct) if present returns true

Removing key-value pairs
    dct.pop(key) removes the item with the specified key name:
    dct.popitem() # removes the last item
    del dct['key2'] # removes key2 item

Changing a dict to list items
    dct.items() method changes dictionary to a list of tuples

Clearing items in a dict
    dct.clear()

Deleting the dict
    del dct

Copying a dict
    using copy() we can avoid mutation of the org dict
    dct = {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}
    dct_copy = dct.copy() # {'key1':'value1', 'key2':'value2', 'key3':'value3', 'key4':'value4'}

Getting Dictionary Keys as a List
    keys = dct.keys() #keys() method gives us all the keys of a a dictionary as a list

Getting Dictionary Values as a List
    values = dct.values() #values() method gives us all the values of a a dictionary as a list
